```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
img = cv2.imread("cyprus_gc.jpg")
img_original = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)/255.0
```

```python
plt.imshow(img_original)
```

```python
img_corrected = img_original ** 1.3
plt.imshow(img_corrected)
```

```python
img_bright = img_original + 0.45
```

```python
plt.imshow(img_bright)
```

```python
img = cv2.imread("cyprus_gc.jpg")
img_original = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
hist, bins = np.histogram(img_bright.ravel(), bins=256, range=(0,255))
plt.plot(bins[0:-1]+0.5, hist), plt.grid(True)
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
```

```python
plt.imshow(clahe.apply(img_original), cmap='gray', vmin=0, vmax=255)
```
